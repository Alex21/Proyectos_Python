```python
#Formato de encabezdo de la boleta
print("=" * 40)
print("  CALCULADORA DE PROPINAS Y CUENTA")
print("=" * 40)
print()
#Parte 1- Creamos el input para el usuario
nombre_restaurante = input("Nombre del restaurante: ")
cuenta_total = input("¿Cuánto fue la cuenta total? S/. ")
porcentaje_propina = input("¿Qué porcentaje de propina quieres dejar? (ej: 10, 15, 20): ")
num_personas = input("¿Entre cuántas personas se dividirá? ")
#Parte 2- Convertimos los datos a numeros
porcentaje_propina= float(porcentaje_propina)
cuenta_total= float(cuenta_total) 
num_personas= int(num_personas)
#Calcular la propina
propina = cuenta_total * (porcentaje_propina / 100)
#Pago de IGV
igv = 0.18 * cuenta_total
#Calcular el total
total_con_propina = cuenta_total + propina + igv
# Pago por persona
pago_por_persona = total_con_propina / num_personas
# MOSTRAR RESULTADOS
print()
print("=" * 40)
print("            RESUMEN DE LA CUENTA")
print("=" * 40)
print("Restaurante: ", nombre_restaurante)
print()
print("Cuenta:                 S/.", cuenta_total)
print("IGV (18%):              S/.", round(igv,2))
print("Propina (" + str(porcentaje_propina) + "%):         S/.", round(propina,2))
print("-" * 40)
print("TOTAL CON PROPINA:      S/.", round(total_con_propina, 2))
print()
print("Personas:              ", num_personas)
print("Cada persona paga:      S/.", round(pago_por_persona, 2))
print("=" * 40)
```
